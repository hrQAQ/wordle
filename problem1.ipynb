```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import json
import statsmodels.graphics.tsaplots as tsaplots
from statsmodels.tsa.stattools import adfuller
import seaborn as sns

# DIRS
DATA_DIR = "data"
IMG_DIR = "img"
WORDLE_DATA_FILE = os.path.join(DATA_DIR, "Problem_C_Data_Wordle.xlsx")
PROD_DATA_FILE = os.path.join(DATA_DIR, "data.xlsx")
# Color pallete for plotting
color_pal = ["#F8766D", "#D39200", "#93AA00",
             "#00BA38", "#00C19F", "#00B9E3",
             "#619CFF", "#DB72FB"]
# plt.style.use('ggplot')
# plt.style.use('seaborn')
plt.style.use('fivethirtyeight')

df = pd.read_excel(WORDLE_DATA_FILE)
df = df[['Date', 'Total-num-repo']]
df.columns = ['Date','report number']
df = df.sort_values(by='Date', ascending=True)
df.reset_index(drop=True, inplace=True)
df.set_index('Date').plot(style='.', figsize=(15,5), color=color_pal[0], title='Primer Data')
plt.figure(dpi=150)
plt.show()
```

我们将创造一些时间序列特征，看看趋势是如何受到星期、季度、月度等影响的。

```python
def create_features(df, label=None):
    """
    Creates time series features from datetime index.
    """
    df = df.copy()
    df['date'] = df.index
    df['dayofweek'] = df['Date'].dt.dayofweek
    df['quarter'] = df['Date'].dt.quarter
    df['month'] = df['Date'].dt.month
    df['dayofyear'] = df['Date'].dt.dayofyear
    df['dayofmonth'] = df['Date'].dt.day
    df['weekofyear'] = df['Date'].dt.isocalendar().week
    
    X = df[['dayofweek','quarter','month',
           'dayofyear','dayofmonth','weekofyear']]
    if label:
        y = df[label]
        return X, y
    return X

df = pd.read_excel(PROD_DATA_FILE)
df = df[['Date', 'Total-num-repo']]
df = df.sort_values(by='Date', ascending=True)
df.reset_index(drop=True, inplace=True)
X, y = create_features(df, label='Total-num-repo')

features_and_target = pd.concat([X, y], axis=1)
print(features_and_target)
```

**Plotting the Features to see trends**

```python
sns.pairplot(features_and_target.dropna(),
             hue='month',
             palette="Set2_r",
             x_vars=['month',
                     'quarter','weekofyear'],
             y_vars='Total-num-repo',
             height=5,
             plot_kws={'alpha':0.5, 'linewidth':0},
            )
plt.suptitle('Reports Number Trend')
plt.show()
```

- 对数变换

> 作用：
>
> 1.数据平稳化：对于具有明显趋势或周期性的数据，进行对数变换可以将其转换为平稳的数据序列。平稳的数据序列更容易建立预测模型，且能够避免一些常见的时间序列分析问题，如异方差性和非常数波动。 
>
> 2.变换为线性关系：对于非线性关系的数据，进行对数变换可以将其转换为线性关系，使得建立预测模型更加简单。例如，对于指数增长的数据，进行对数变换可以将其转换为线性增长，使得可以使用线性回归模型进行分析
>
> 3.改善预测精度：对数变换可以将数据的比例关系转换为差异关系，从而能够更好地处理数据中的离群值或极端值。同时，对于呈现指数增长或减少趋势的数据，进行对数变换可以降低数据的波动性，从而使得预测结果更加准确。
>
> 总之，对数变换是一种常见的数据预处理方法，能够帮助我们更好地理解和处理数据，以及建立更加准确的预测模型。但是需要注意的是，对数变换可能会对数据的实际含义造成一定的影响，因此在使用时需要谨慎考虑。

#### 模型选择

时间序列预测模型可以考虑使用ARIMA、prophet等。

Prophet模型和ARIMA模型都是用于时间序列预测的常用模型，它们各有优势，下面是一些Prophet模型相对于ARIMA模型的优势：

- 容易使用：Prophet模型相对于ARIMA模型来说更加易于使用，不需要太多的数学背景知识和专业技能，可以更快速地进行建模和预测。

- 更好的处理季节性和节假日效应：Prophet模型专门设计用于处理具有强烈季节性和节假日效应的时间序列数据，而ARIMA模型则需要手动指定这些效应。在处理复杂的时间序列数据时，Prophet模型可以更准确地捕捉和预测这些效应。

- 更好的处理缺失值和异常值：Prophet模型可以在数据中包含缺失值的情况下进行建模和预测，而ARIMA模型需要对缺失值进行额外的处理。此外，Prophet模型还可以更好地处理异常值，并通过指定一个阈值来控制异常值的影响。

- 更好的处理非线性趋势：Prophet模型可以捕捉到非线性的趋势，而ARIMA模型通常只能捕捉线性的趋势。这使得Prophet模型更适合处理具有复杂趋势的时间序列数据。

总之，Prophet模型和ARIMA模型各有优势，在实际使用中需要根据数据的特点选择合适的模型。

先来看Prophet-Normal模型

```python
from prophet import Prophet
train_df.columns = ["ds", "y"]
test_df.columns = ["ds", "y"]
m1 = Prophet(interval_width=0.7)
m1.fit(train_df)
future1 = m1.make_future_dataframe(periods=60)
forecast1 = m1.predict(future1)
forecast1[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
m1.plot(forecast1)
plt.show()
```

下面是趋势图和每周的趋势

```python
m1.plot_components(forecast1)
plt.show()
```

下面是预测结果

```python
predict = np.exp(forecast1[['yhat', 'yhat_lower', 'yhat_upper']])
real = np.exp(test_df['y'])
plt.plot(test_df['ds'], real, label='test')
plt.plot(test_df['ds'], predict['yhat'][-31:], label='yhat')
plt.plot(test_df['ds'], predict['yhat_lower'][-31:], label='yhat_lower')
plt.plot(test_df['ds'], predict['yhat_upper'][-31:], label='yhat_upper')
plt.tight_layout()
# x轴标签倾斜
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

**Normal模型误差分析**

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
pred = predict['yhat'][-31:]
MSE = mean_squared_error(y_true=real,y_pred=pred) # MSE
MAE = mean_absolute_error(y_true=real,y_pred=pred) # MAE
def mean_absolute_percentage_error(y_true, y_pred): 
    """Calculates MAPE given y_true and y_pred"""
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
MAPE = mean_absolute_percentage_error(y_true=real,y_pred=pred)
print("MSE: ", MSE)
print("MAE: ", MAE)
print("MAPE: ", MAPE,"%")

```
