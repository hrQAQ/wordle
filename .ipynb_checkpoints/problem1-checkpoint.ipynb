我们将创造一些时间序列特征，看看趋势是如何受到星期、季度、月度等影响的。

```python
def create_features(df, label=None):
    """
    Creates time series features from datetime index.
    """
    df = df.copy()
    df['date'] = df.index
    df['dayofweek'] = df['Date'].dt.dayofweek
    df['quarter'] = df['Date'].dt.quarter
    df['month'] = df['Date'].dt.month
    df['dayofyear'] = df['Date'].dt.dayofyear
    df['dayofmonth'] = df['Date'].dt.day
    df['weekofyear'] = df['Date'].dt.isocalendar().week
    
    X = df[['dayofweek','quarter','month',
           'dayofyear','dayofmonth','weekofyear']]
    if label:
        y = df[label]
        return X, y
    return X

df = pd.read_excel(PROD_DATA_FILE)
df = df[['Date', 'Total-num-repo']]
df = df.sort_values(by='Date', ascending=True)
df.reset_index(drop=True, inplace=True)
X, y = create_features(df, label='Total-num-repo')

features_and_target = pd.concat([X, y], axis=1)
print(features_and_target)
```

- 对数变换

> 作用：
>
> 1.数据平稳化：对于具有明显趋势或周期性的数据，进行对数变换可以将其转换为平稳的数据序列。平稳的数据序列更容易建立预测模型，且能够避免一些常见的时间序列分析问题，如异方差性和非常数波动。 
>
> 2.变换为线性关系：对于非线性关系的数据，进行对数变换可以将其转换为线性关系，使得建立预测模型更加简单。例如，对于指数增长的数据，进行对数变换可以将其转换为线性增长，使得可以使用线性回归模型进行分析
>
> 3.改善预测精度：对数变换可以将数据的比例关系转换为差异关系，从而能够更好地处理数据中的离群值或极端值。同时，对于呈现指数增长或减少趋势的数据，进行对数变换可以降低数据的波动性，从而使得预测结果更加准确。
>
> 总之，对数变换是一种常见的数据预处理方法，能够帮助我们更好地理解和处理数据，以及建立更加准确的预测模型。但是需要注意的是，对数变换可能会对数据的实际含义造成一定的影响，因此在使用时需要谨慎考虑。

#### 模型选择

时间序列预测模型可以考虑使用ARIMA、prophet等。

Prophet模型和ARIMA模型都是用于时间序列预测的常用模型，它们各有优势，下面是一些Prophet模型相对于ARIMA模型的优势：

- 容易使用：Prophet模型相对于ARIMA模型来说更加易于使用，不需要太多的数学背景知识和专业技能，可以更快速地进行建模和预测。

- 更好的处理季节性和节假日效应：Prophet模型专门设计用于处理具有强烈季节性和节假日效应的时间序列数据，而ARIMA模型则需要手动指定这些效应。在处理复杂的时间序列数据时，Prophet模型可以更准确地捕捉和预测这些效应。

- 更好的处理缺失值和异常值：Prophet模型可以在数据中包含缺失值的情况下进行建模和预测，而ARIMA模型需要对缺失值进行额外的处理。此外，Prophet模型还可以更好地处理异常值，并通过指定一个阈值来控制异常值的影响。

- 更好的处理非线性趋势：Prophet模型可以捕捉到非线性的趋势，而ARIMA模型通常只能捕捉线性的趋势。这使得Prophet模型更适合处理具有复杂趋势的时间序列数据。

总之，Prophet模型和ARIMA模型各有优势，在实际使用中需要根据数据的特点选择合适的模型。

先来看Prophet-Normal模型

```python
from prophet import Prophet
train_df.columns = ["ds", "y"]
test_df.columns = ["ds", "y"]
m1 = Prophet(interval_width=0.7)
m1.fit(train_df)
future1 = m1.make_future_dataframe(periods=60)
forecast1 = m1.predict(future1)
forecast1[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
m1.plot(forecast1)
plt.show()
```

下面是趋势图和每周的趋势

```python
m1.plot_components(forecast1)
plt.show()
```

下面是预测结果

```python
predict = np.exp(forecast1[['yhat', 'yhat_lower', 'yhat_upper']])
real = np.exp(test_df['y'])
plt.plot(test_df['ds'], real, label='test')
plt.plot(test_df['ds'], predict['yhat'][-31:], label='yhat')
plt.plot(test_df['ds'], predict['yhat_lower'][-31:], label='yhat_lower')
plt.plot(test_df['ds'], predict['yhat_upper'][-31:], label='yhat_upper')
plt.tight_layout()
# x轴标签倾斜
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

**Normal模型误差分析**

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
pred = predict['yhat'][-31:]
MSE = mean_squared_error(y_true=real,y_pred=pred) # MSE
MAE = mean_absolute_error(y_true=real,y_pred=pred) # MAE
def mean_absolute_percentage_error(y_true, y_pred): 
    """Calculates MAPE given y_true and y_pred"""
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
MAPE = mean_absolute_percentage_error(y_true=real,y_pred=pred)
print("MSE: ", MSE)
print("MAE: ", MAE)
print("MAPE: ", MAPE,"%")

```

prepare 2. **提取特征**

导入已经预处理好的特征表

```python
K_MEAN_FILE = os.path.join(DATA_DIR, "k_means.json")
POSS_WORDS_FREQ =  os.path.join(DATA_DIR, "possible_words_freq.json")
POSS_WORDS_FREQ_SIGMOID = os.path.join(DATA_DIR, "possible_words_sigmoid.json")
POSS_WORDS_FILE = os.path.join(DATA_DIR, "possible_words.txt")
AEIOU = os.path.join(DATA_DIR, "aeiou.json")
FIRST_LETTER = os.path.join(DATA_DIR, "first_letter.json")
k_mean = json.load(open(K_MEAN_FILE))
freq = json.load(open(POSS_WORDS_FREQ))
freq_sigmoid = json.load(open(POSS_WORDS_FREQ_SIGMOID))
aeiou = json.load(open(AEIOU))
first_letter = json.load(open(FIRST_LETTER))

def centerize(x):
    """
    对向量x进行中心化处理，即减去均值。
    
    参数：
    x: numpy array，需要进行中心化处理的向量
    
    返回：
    中心化后的向量
    """
    mean_x = np.mean(x)
    centered_x = x - mean_x
    return centered_x

def centerize_matrix(X):
    """
    对矩阵X的每一列进行中心化处理，即减去每一列的均值。
    
    参数：
    X: numpy array，需要进行中心化处理的矩阵
    
    返回：
    中心化后的矩阵
    """
    mean_X = np.mean(X, axis=0)
    centered_X = X - mean_X
    return centered_X

possibility = centerize(possibility)
```

1. 对K_means**单词编辑距离聚类**特征进行相关性分析

```python
import scipy.stats as stats
words_k_mean = centerize([k_mean[word] for word in words])
corr, pval = stats.pearsonr(words_k_mean, possibility)
# 输出Pearson相关系数和p值
print("words_k_mean Pearson correlation coefficient: ", corr)
print("p-value: ", pval)
```

Pearson相关系数为0.047，接近于0，说明两个变量之间没有强烈的线性关系。这意味着当一个变量变化时，另一个变量不一定会随之变化。

p值为0.369，大于通常用于确定统计显着性的阈值0.05，这意味着相关系数不够显著，无法排除由于偶然因素导致观察到的相关性。因此，无法确定两个变量之间是否存在显著的线性关系。

需要注意的是，即使Pearson相关系数接近于0，仍然有可能存在非线性关系。此外，Pearson相关系数只能够检测两个变量之间的线性关系，如果存在其他类型的关系，如指数关系或幂关系，Pearson相关系数可能不适用。因此，在解释相关系数之前，需要仔细考虑数据的特征和相关性分析方法的适用性。

5. 对**重音位置**特征进行相关性分析

使用自然语言处理工具包nltk，提取单词的重音位置

```python
from nltk.corpus import cmudict

def get_stress(word):
    d = cmudict.dict()
    if word.lower() in d:
        # 返回重音数字，0表示无重音，1、2、3分别表示第一、第二、第三个音节重音
        return [len(list(y for y in x if y[-1].isdigit())) for x in d[word.lower()]][0]
    else:
        return 0
    
get_stress("pronunciation")

words_stress = centerize([get_stress(word) for word in words])
# print(words_stress)
corr, pval = stats.pearsonr(words_stress, possibility)
# 输出Pearson相关系数和p值
print("words_stress Pearson correlation coefficient: ", corr)
print("p-value: ", pval)
```

6. 对**词性**特征进行相关性分析

使用自然语言处理工具包nltk，提取单词的词根，有词根的单词标记为1，没有词根的单词标记为0。

```python
import nltk

# nltk.download('punkt')
# nltk.download('averaged_perceptron_tagger')

def get_pos(word):
    """
    获取单词的词性
    :param word: 需要获取词性的单词
    :return: 单词的词性
    """
    tokens = nltk.word_tokenize(word)
    pos_tags = nltk.pos_tag(tokens)
    return pos_tags[0][1]

# 将词性标记映射为数值
pos_tag_dict = {
    'CC': 1, 'CD': 2, 'DT': 3, 'EX': 4, 'FW': 5, 'IN': 6, 'JJ': 7, 'JJR': 8, 'JJS': 9, 'LS': 10, 
    'MD': 11, 'NN': 12, 'NNS': 13, 'NNP': 14, 'NNPS': 15, 'PDT': 16, 'POS': 17, 'PRP': 18, 'PRP$': 19, 
    'RB': 20, 'RBR': 21, 'RBS': 22, 'RP': 23, 'SYM': 24, 'TO': 25, 'UH': 26, 'VB': 27, 'VBD': 28, 
    'VBG': 29, 'VBN': 30, 'VBP': 31, 'VBZ': 32, 'WDT': 33, 'WP': 34, 'WP$': 35, 'WRB': 36
}
    
    
# 测试代码
word = "cat"
print(pos_tag_dict[get_pos(word)]) # 输出NN

word = "done"
print(pos_tag_dict[get_pos(word)]) # 输出VB

words_pos = centerize([pos_tag_dict[get_pos(word)] for word in words])
corr, pval = stats.pearsonr(words_pos, possibility)
# 输出Pearson相关系数和p值
print("words_pos Pearson correlation coefficient: ", corr)
print("p-value: ", pval)
```

根据给出的结果，words_pos和Pearson相关系数的值为0.038，接近于0，说明两个变量之间的线性关系很弱。p-value的值为0.469，大于0.05，说明相关系数的统计学显著性不强。综合来看，words_pos与Pearson相关系数之间没有强的线性关系，而且相关系数的显著性不强。

7. 对**词根**特征进行相关性分析

使用自然语言处理工具包nltk，提取单词的词根，有词根的单词标记为1，没有词根的单词标记为0。

```python
from nltk.corpus import wordnet as wn

def has_root(word):
    """
    判断单词是否具有词根
    :param word: 需要判断的单词
    :return: 布尔值，如果单词具有词根则返回True，否则返回False
    """
    synsets = wn.synsets(word)
    return len(synsets) > 0 and len(synsets[0].lemmas()) > 0 and synsets[0].lemmas()[0].name() != word

word = "doing"
print(has_root(word)) # 输出True

word = "apple"
print(has_root(word)) # 输出False

# 判断所有单词是否具有词根, 有词根的单词标记为1，否则标记为0
words_has_root = centerize([1 if has_root(word) else 0 for word in words])
corr, pval = stats.pearsonr(words_has_root, possibility)
# 输出Pearson相关系数和p值
print("words_has_root Pearson correlation coefficient: ", corr)
print("p-value: ", pval)
```

Pearson相关系数为-0.03247377515247207，接近于0，说明两个变量之间没有强烈的线性关系。这意味着当一个变量变化时，另一个变量不一定会随之变化。

p值为0.5396739152233343，大于通常用于确定统计显着性的阈值0.05，这意味着相关系数不够显著，无法排除由于偶然因素导致观察到的相关性。因此，无法确定两个变量之间是否存在显著的线性关系。

8. 对**情感**特征进行相关性分析

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

def get_word_sentiment(word):
    '''
    获取单词的情感倾向，返回值介于-1和1之间，值越大表示情感越积极
    '''
    sia = SentimentIntensityAnalyzer()
    score = sia.polarity_scores(word)
    return score['compound']


# 测试代码
word = "happy"
print(get_word_sentiment(word)) # 输出0.8，表示单词"happy"情感积极

word = "sad"
print(get_word_sentiment(word)) # 输出-0.5，表示单词"sad"情感消极

words_sentiment = centerize([get_word_sentiment(word) for word in words])
# 输出Pearson相关系数和p值
corr, pval = stats.pearsonr(words_sentiment, possibility)
print("words_sentiment Pearson correlation coefficient: ", corr)
print("p-value: ", pval)
```

前面的分析主要是在使用Python中的Scipy库计算两个变量之间的皮尔逊相关系数和p值，以研究这两个变量之间的相关性。具体来说，我们分别计算了单词频率（words_freq）和单词情感分析结果（words_sentiment）之间的皮尔逊相关系数，发现它们之间的相关性程度都比较低。这意味着单词的使用频率和单词的情感极性之间并没有非常明显的关联。但是，需要注意的是，这种相关性并不能完全反映语境和个人主观感受的影响，因此具体情况可能会因人而异。

9. 对**字符频率加权**特征进行相关性分析

```python
freq_dict = {'e': 12.70, 't': 9.06, 'a': 8.17, 'o': 7.51, 'i': 6.97, 'n': 6.75, 's': 6.33, 'h': 6.09, 'r': 5.99, 'd': 4.25, 'l': 4.03, 'c': 2.78, 'u': 2.76, 'm': 2.41, 'w': 2.36, 'f': 2.23, 'g': 2.02, 'y': 1.97, 'p': 1.93, 'b': 1.49, 'v': 0.98, 'k': 0.77, 'j': 0.15, 'x': 0.15, 'q': 0.10, 'z': 0.07}


def get_weight(word):
    weight = 0
    for char in word:
        weight += freq_dict[char]
    return weight
    
words_char_freq = centerize([get_weight(word) for word in words])
# 输出Pearson相关系数和p值
corr, pval = stats.pearsonr(words_char_freq, possibility)
print("words_char_freq Pearson correlation coefficient: ", corr)
print("p-value: ", pval)
```

在这种情况下，Pearson相关系数为0.0162，这意味着英文单词的长度与其字母出现频率之间存在非常微弱的正相关关系。p-value是对Pearson相关系数假设检验的结果，它表示当假设“英文单词的长度与其字母出现频率之间不存在任何相关性”时，通过随机采样得到的Pearson相关系数或更极端情况出现的概率。在这种情况下，p-value为0.7595，这表明我们不能拒绝原假设，即英文单词的长度与其字母出现频率之间不存在任何相关性的假设。

10. 对**Word2Vec**特征进行相关性分析

```python
from gensim.models import Word2Vec
import nltk

sentences = nltk.corpus.brown.sents()

# 训练模型
model = Word2Vec(sentences, window=5, min_count=0, workers=4)
print("model prepared")
# 查看单词向量
vector = model.wv['dog']

def get_word_vectors(words, model):
    word_vectors = []
    for word in words:
        if word in model.wv.key_to_index:
            word_vectors.append(model.wv[word])
        else:
            similar_word = model.wv.index_to_key[model.wv.most_similar(word, topn=1)[0][0]]
            word_vectors.append(model.wv[similar_word])
    return np.array(word_vectors)

words_vector = get_word_vectors(words, model)
print("corr cal start")
# 计算相关性向量和p-value向量
corr_vector = []
pval_vector = []
for i in range(len(words_vector[0])):
    corr, pval = stats.pearsonr([word_vector[i] for word_vector in words_vector], possibility)
    corr_vector.append(corr)
    pval_vector.append(pval)
for i in range(len(corr_vector)):
    print("word_vector ", i, " Pearson correlation coefficient: ", corr_vector[i])
    print("p-value: ", pval_vector[i])

```
